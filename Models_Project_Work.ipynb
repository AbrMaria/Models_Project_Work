```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
est = smf.ols('TARGET ~ var15 + var3 + var36 + var21 + var38', data_train).fit()
est_res = est.summary()
est_res.tables[1]
```

```python
#все выбранные показатели, кроме var3 и var21, имеют маленький  p-value. 
#Т.е. между var3, var21 и TARGET нет линейной зависимости и, следовательно, от этих атрибутов можно избавиться.
```

```python
est = smf.ols('TARGET ~ var15 + var36 + var38', data_train).fit()
est_res = est.summary()
est_res.tables[1]
```

```python
three_x_lm = smf.ols('TARGET ~ var15 + var36 + var38', data_train).fit()
rss = np.sum(three_x_lm.resid ** 2)
print("RSS:", rss)
print("RSE:", np.sqrt(rss / (data_train.shape[0] - 3 - 1)))
print("R^2:", three_x_lm.rsquared)
```

```python
two_x_lm = smf.ols('TARGET ~ var15 + var36', data_train).fit()
rss = np.sum(two_x_lm.resid ** 2)
print("RSS:", rss)
print("RSE:", np.sqrt(rss / (data_train.shape[0] - 2 - 1)))
print("R^2:", two_x_lm.rsquared)
```
